```python
#Eslam Nasser 20180047
#Farrag mohamed farrag 20180197

import numpy as np

def sigmoid(x):
  # Sigmoid activation function: f(x) = 1 / (1 + e^(-x))
  return 1 / (1 + np.exp(-x))

def deriv_sigmoid(x):
  # Derivative of sigmoid: f'(x) = f(x) * (1 - f(x))
  fx = sigmoid(x)
  return fx * (1 - fx)

def mse_loss(y_true, y_pred):
  # y_true and y_pred are numpy arrays of the same length.
  return ((y_true - y_pred) ** 2).mean()
class OurNeuralNetwork:
  '''
  A neural network with:
    - 2 inputs
    - a hidden layer with 4 neurons (neuron1, neuron2,neuron3,neuron4)
    - an output layer with 1 neuron (output)
    
    '''

  def __init__(self):
    # Weights
    self.Weights1 = np.random.normal()
    self.Weights2 = np.random.normal()
    self.Weights3 = np.random.normal()
    self.Weights4 = np.random.normal()
    self.Weights5 = np.random.normal()
    self.Weights6 = np.random.normal()
    self.Weights7 = np.random.normal()
    self.Weights8 = np.random.normal()
    self.Weights9 = np.random.normal()
    self.Weights10 = np.random.normal()
    self.Weights11 = np.random.normal()
    self.Weights12 = np.random.normal()
    # Biases
    self.biase1 = np.random.normal()
    self.biase2 = np.random.normal()
    self.biase3 = np.random.normal()
    self.biase4 = np.random.normal()
    self.biase5 = np.random.normal()

  def feedforward(self, x):
    # x is a numpy array with 4 elements.
    neuron1 = sigmoid(self.Weights1 * x[0] + self.Weights2 * x[1] + self.biase1)
    neuron2 = sigmoid(self.Weights3 * x[0] + self.Weights4 * x[1] + self.biase2)
    neuron3 = sigmoid(self.Weights5 * x[0] + self.Weights6 * x[1] + self.biase3)
    neuron4 = sigmoid(self.Weights7 * x[0] + self.Weights8 * x[1] + self.biase4)
    output = sigmoid(self.Weights9 * neuron1 + self.Weights10 * neuron2 +self.Weights11 * neuron3+self.Weights12 * neuron4+ self.biase5)
    return output

  def train(self, data, y_trues):
    
    learn_rate =0.005
    epochs = 100 # number of times to loop through the entire dataset

    for epoch in range(epochs):
      for x, y_true in zip(data, y_trues):
        # --- Do a feedforward (we'll need these values later)
        sum_of_neuron1 = self.Weights1 * x[0] + self.Weights2 * x[1] + self.biase1
        neuron1 = sigmoid(sum_of_neuron1)

        sum_of_neuron2 = self.Weights3 * x[0] + self.Weights4 * x[1] + self.biase2
        neuron2 = sigmoid(sum_of_neuron2)
        
        sum_of_neuron3 = self.Weights5 * x[0] + self.Weights6 * x[1] + self.biase3
        neuron3 = sigmoid(sum_of_neuron3)
        
        sum_of_neuron4 = self.Weights7 * x[0] + self.Weights8 * x[1] + self.biase4
        neuron4 = sigmoid(sum_of_neuron4)

        sum_of_output = self.Weights9 * neuron1 + self.Weights10 * neuron2 +  self.Weights11 * neuron3+ self.Weights12 * neuron4 + self.biase5
        output = sigmoid(sum_of_output)
        y_pred = output
         # --- Calculate partial derivatives.
        # --- Naming: d_L_d_w1 represents "partial L / partial w1"
        d_L_d_ypred = -2 * (y_true - y_pred)

        # Neuron output
        d_ypred_d_Weights9 = neuron1 * deriv_sigmoid(sum_of_output )
        d_ypred_d_Weights10 = neuron2 * deriv_sigmoid(sum_of_output )
        d_ypred_d_Weights11 = neuron3 * deriv_sigmoid(sum_of_output )
        d_ypred_d_Weights12 = neuron4 * deriv_sigmoid(sum_of_output )
        d_ypred_d_biase5 = deriv_sigmoid(sum_of_output )

        d_ypred_d_neuron1 = self.Weights9 * deriv_sigmoid(sum_of_output )
        d_ypred_d_neuron2 = self.Weights10 * deriv_sigmoid(sum_of_output )
        d_ypred_d_neuron3 = self.Weights11 * deriv_sigmoid(sum_of_output )
        d_ypred_d_neuron4 = self.Weights12 * deriv_sigmoid(sum_of_output )

        # Neuron neuron1
        d_neuron1_d_Weights1 = x[0] * deriv_sigmoid( sum_of_neuron1)
        d_neuron1_d_Weights2 = x[1] * deriv_sigmoid( sum_of_neuron1)
        d_neuron1_d_biase1 = deriv_sigmoid( sum_of_neuron1)

        # Neuron neuron2
        d_neuron2_d_Weights3 = x[0] * deriv_sigmoid( sum_of_neuron2)
        d_neuron2_d_Weights4 = x[1] * deriv_sigmoid(sum_of_neuron2)
        d_neuron2_d_biase2 = deriv_sigmoid( sum_of_neuron2)
        # Neuron neuron3
        d_neuron3_d_Weights5 = x[0] * deriv_sigmoid( sum_of_neuron3)
        d_neuron3_d_Weights6 = x[1] * deriv_sigmoid(sum_of_neuron3)
        d_neuron3_d_biase3 = deriv_sigmoid( sum_of_neuron3)

        # Neuron neuron4
        d_neuron4_d_Weights7 = x[0] * deriv_sigmoid(sum_of_neuron4)
        d_neuron4_d_Weights8 = x[1] * deriv_sigmoid( sum_of_neuron4)
        d_neuron4_d_biase4 = deriv_sigmoid( sum_of_neuron4)
        
        # --- Update weights and biases
        # Neuron neuron1
        self.Weights1 -= learn_rate * d_L_d_ypred * d_ypred_d_neuron1 * d_neuron1_d_Weights1
        self.Weights2 -= learn_rate * d_L_d_ypred * d_ypred_d_neuron1 * d_neuron1_d_Weights2
        self.biase1 -= learn_rate * d_L_d_ypred * d_ypred_d_neuron1 * d_neuron1_d_biase1

        # Neuron neuron2
        self.Weights3 -= learn_rate * d_L_d_ypred * d_ypred_d_neuron2 * d_neuron2_d_Weights3
        self.Weights4 -= learn_rate * d_L_d_ypred * d_ypred_d_neuron2 * d_neuron2_d_Weights4
        self.biase2 -= learn_rate * d_L_d_ypred * d_ypred_d_neuron2 * d_neuron2_d_biase2
        
        # Neuron neuron3
        self.Weights5 -= learn_rate * d_L_d_ypred * d_ypred_d_neuron3 * d_neuron3_d_Weights5
        self.Weights6 -= learn_rate * d_L_d_ypred * d_ypred_d_neuron3 * d_neuron3_d_Weights6
        self.biase3 -= learn_rate * d_L_d_ypred * d_ypred_d_neuron3 * d_neuron3_d_biase3
        
        
        # Neuron neuron4
        self.Weights7 -= learn_rate * d_L_d_ypred * d_ypred_d_neuron4 * d_neuron4_d_Weights7
        self.Weights8 -= learn_rate * d_L_d_ypred * d_ypred_d_neuron4 * d_neuron4_d_Weights8
        self.biase4 -= learn_rate * d_L_d_ypred * d_ypred_d_neuron4 * d_neuron4_d_biase4

        # Neuron output
        self.Weights9 -= learn_rate * d_L_d_ypred * d_ypred_d_Weights9
        self.Weights10 -= learn_rate * d_L_d_ypred * d_ypred_d_Weights10
        self.Weights11 -= learn_rate * d_L_d_ypred * d_ypred_d_Weights11
        self.Weights12 -= learn_rate * d_L_d_ypred * d_ypred_d_Weights12
        self.biase5 -= learn_rate * d_L_d_ypred * d_ypred_d_biase5

         # --- Calculate total loss at the end of each epoch
      if epoch % 1 == 0:
        y_preds = np.apply_along_axis(self.feedforward, 1, data)
        loss = mse_loss(y_trues, y_preds)
        print("Epoch %d loss: %.3f" % (epoch, loss))

# Define dataset
data = np.array([
  [-2, -1],  
  [25, 6],   
  [17, 4],   
  [-15, -6],
  [-5,4],
  [-7, -3],
])
y_trues = np.array([
  1, 
  0, 
  0, 
  1, 
  1,
  1,
])
# Train our neural network!
network = OurNeuralNetwork()
network.train(data, y_trues)
```
